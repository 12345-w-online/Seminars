```python
my_dir_path = './'
```

```python
# итератор содержимого папки
entries = os.scandir(my_dir_path)
entries
```

```python
# проход по файлам в заданной папке
with os.scandir(my_dir_path) as entries:
    for entry in entries:
        print(entry.name)
```

```python
# alternative way to scan through the folder
from pathlib import Path

entries = Path(my_dir_path)
for entry in entries.iterdir():
    print(entry.name)
```

```python
# static list of folder contents
os.listdir(my_dir_path)
```

```python
# list FILES and DIRs separately
with os.scandir(my_dir_path) as entries:
    for entry in entries:
        if entry.is_file(): 
            print(f"file: {entry.name}")
        if entry.is_dir():
            print(f"dir: {entry.name}")
```

```python
# create new folder
os.makedirs('./example_directory/subfolder/subsubfolder', 
            exist_ok=True)
```

```python
# get .txt files only
for f_name in os.listdir(my_dir_path):
    if f_name.endswith('.mp4'):
        print(f_name)
```

```python
# Walk through a directory tree and print the names of the directories and files
for dirpath, dirnames, files in os.walk(my_dir_path, 
                                        topdown=False # if you want bottom-up listing
                                       ):
    print(f'Found directory: {dirpath} with name {dirnames}')
    
    print('')
    for file_name in files:
        print(file_name)
```

```python
# remove file
data_file_path = './useless_file.txt'

# option 1
if os.path.isfile(data_file_path):
    os.remove(data_file_path)

# option 2
if os.path.isfile(data_file_path):
    os.unlink(data_file_path)
```

```python
# delete a signle empty directory
trash_dir = './bad_dir'

try:
    os.rmdir(trash_dir)
except OSError as e:
    print(f'Error: {trash_dir} : {e.strerror}')
```

```python
# delete directory tree (any dir tree)
import shutil

trash_dir = 'my_documents/bad_dir'

try:
    shutil.rmtree(trash_dir)
except OSError as e:
    print(f'Error: {trash_dir} : {e.strerror}\n {e}')
```
