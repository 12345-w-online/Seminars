1. Задан неориентированный граф. Определите является ли граф связным,  т. е. между любой парой вершин этого графа существует как минимум один путь.

```python
import numpy as np
import typing


def dfs(data: dict, start_el: typing.Union[int, str]) -> set:  # Вовращает список посещенных элем. из выбранного начала
    flag = True
    elem = start_el
    stack = []
    visited = set(start_el)
    while flag:
        edges = set(data[elem])
        edges = sorted(list(edges.difference(visited)))
        # print(edges)
        if not edges:
            if not stack:
                visited.add(str(elem))
                flag = False
            else:
                elem = stack[-1]
                stack = stack[0: len(stack) - 1]
        else:
            visited.add(str(edges[0]))
            stack = stack + list(reversed(edges[1::]))
            elem = edges[0]
    return visited


def update_dict(data: dict, data_key: typing.Union[int, str], add_value: typing.Union[int, str]):
    if str(data_key) in data:
        set_for_key = data[str(data_key)]
    else:
        set_for_key = set()
    set_for_key = set_for_key.union(set(str(add_value)))
    data.update({str(data_key): set_for_key})
    return data


def create_data_dict(num: int, directed=False) -> dict:
    data = dict()
    for _ in range(num):
        edge = tuple(map(int, input().split()))
        data = update_dict(data, edge[0], edge[1])
        if not directed:
            data = update_dict(data, edge[1], edge[0])
    return data


def create_data_matrix(n) -> np.ndarray:
    matrix = np.zeros((n + 1, n + 1))
    matrix[(0, 0)] = None
    for _ in range(1, n + 1):
        matrix[(_, 0)] = _
        matrix[(0, _)] = _
    for _ in range(n):
        edge = tuple(map(int, input().split()))
        matrix[edge] = 1
        matrix[edge[::-1]] = 1
    return matrix


n = int(input())
data = create_data_dict(n)
visited = dfs(data, "0")
vertex = set(data.keys())
# print(f"vertex {vertex}")
# print(f"visited {visited}")
# print(vertex.difference(visited))
if not vertex.difference(visited):
    print("True")
else:
    print("False")

'''
Test:
5
0 1
1 2
2 3
3 5
4 1
7
0 1
1 2
2 3
3 5
4 6
4 8
7 8
'''
```
