```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('iris_data.csv')
A = list(df['PetalLengthCm'])
data_1 = 0
data_2 = 0
data_3 = 0 
for i in A:
    if float(i) > 1.2:
        data_1 += 1
    if float(i) > 1.2 and float(i) < 1.5:
        data_2 += 1
    if float(i) > 1.5:
        data_3 +=1
```

```python
plt.pie([data_1 , data_2, data_3], labels = ['x> 1.2','1.5> x >1.2','x>1.5'])
plt.title('PetalLengthCm')
plt.show()
```
