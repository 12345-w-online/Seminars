```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))
ax1 = plt.subplot2grid((3, 2), (0, 0), colspan=3)
ax2 = plt.subplot2grid((3, 2), (1, 0), colspan=3)
ax1.hist(np.random.normal(0, 2, 1000), 100)
ax2.hist(np.random.normal(0, 2, 100000), 100)
plt.show
```
