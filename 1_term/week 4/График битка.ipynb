```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('BTC_data.csv')
data_x = data['time']
data_y = data['close']
```

```python
ax = plt.figure(figsize=(15,5))
ax = plt.plot(data_x, data_y)
ax = plt.title('Биток')
ax = plt.xlabel('Даты')
ax = plt.ylabel('Деньги')
ax = plt.xticks([data_x[i] for i in range(0, len(data_x), int(len(data_x)/4))])
```
